```python
import pandas as pd

# Ruta del archivo CSV
ruta_csv = "../datos/02. Datos tratamiento preliminar/01_datos_tratamiento_preliminar - V2.csv"

# Intentar leer con separador ';' (muy común en CSV exportados desde Excel)
datos = pd.read_csv(ruta_csv, sep=';', engine='python', on_bad_lines='skip')

# Mostrar las primeras filas
print(datos)
```

# Análisis por origen solicitud

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar CSV
#df = pd.read_csv("../datos/02. Datos tratamiento preliminar/01_datos_tratamiento_preliminar - V2.csvv", sep=';', engine='python')

# Crear columna combinada
datos['Tipo_origen'] = datos.apply(lambda row: 'Solicitud de Información' if row['PL_ORIGEN_DE_SOLICITUD'] == True else 'Solicitud de Admisión', axis=1)

# Pie chart general
origen_pct = datos['Tipo_origen'].value_counts(normalize=True) * 100
plt.figure(figsize=(6,6))
plt.pie(origen_pct, labels=origen_pct.index, autopct='%1.1f%%', startangle=90, colors=["#00FF00", "#0077FF"])
plt.title('Distribución Nacional vs Internacional')
plt.axis('equal')
plt.show()

# Barras por target
conteo_origen_target = datos.groupby(['Tipo_origen', 'target']).size().unstack(fill_value=0)
conteo_origen_target.plot(kind='bar', stacked=False, figsize=(8,5), color=['#00FF00', '#0077FF'])
plt.title('Nacional vs Internacional vs Estado de Matrícula')
plt.xlabel('Tipo de origen')
plt.ylabel('Cantidad de personas')
plt.legend(['No matriculado (target=0)', 'Matriculado (target=1)'])
plt.show()
```
