```python
import pandas as pd

# Ruta del archivo Excel
rs = 'DataSET_SF.xlsx'

# Abrir el archivo Excel
excel_file = pd.ExcelFile(rs)

# Número total de hojas
num_hojas = len(excel_file.sheet_names)
print(f"Número total de hojas: {num_hojas}\n")

# Recorrer cada hoja
for hoja in excel_file.sheet_names:
    df = pd.read_excel(rs, sheet_name=hoja)
    filas, columnas = df.shape
    print(f"\nHoja: {hoja} -> Filas: {filas}, Columnas: {columnas}")
    
    # Calcular % de valores NA por columna
    porcentaje_na = df.isnull().mean() * 100
    
    # Mostrar ordenado de mayor a menor sin truncar
    print("Porcentaje de valores NA por columna:")
    print(porcentaje_na.sort_values(ascending=False).to_string())
```

#### Número de columnas por porcentajes

```python
import pandas as pd

# Ruta del archivo Excel
rs = 'DataSET_SF.xlsx'
excel_file = pd.ExcelFile(rs)

# Número total de hojas
num_hojas = len(excel_file.sheet_names)
print(f"Número total de hojas: {num_hojas}\n")

# Recorrer cada hoja
for hoja in excel_file.sheet_names:
    df = pd.read_excel(rs, sheet_name=hoja)
    filas, columnas = df.shape
    print(f"\nHoja: {hoja} -> Filas: {filas}, Columnas: {columnas}")
    
    # Calcular % de valores NA por columna
    porcentaje_na = df.isnull().mean() * 100
    
    # Contar columnas por rangos de % NA
    rangos = {
        "0% NA": ((porcentaje_na == 0).sum()),
        "1-25% NA": (((porcentaje_na > 0) & (porcentaje_na <= 25)).sum()),
        "26-50% NA": (((porcentaje_na > 25) & (porcentaje_na <= 50)).sum()),
        "51-75% NA": (((porcentaje_na > 50) & (porcentaje_na <= 75)).sum()),
        "76-99% NA": (((porcentaje_na > 75) & (porcentaje_na < 100)).sum()),
        "100% NA": ((porcentaje_na == 100).sum())
    }
    
    print("Número de columnas por rango de % NA:")
    for rango, cantidad in rangos.items():
        print(f"{rango}: {cantidad}")
```

## Porcentaje de nulos, vacíos, etc....

```python
import pandas as pd

# Ruta del archivo Excel
rs = 'DataSET_SF.xlsx'
excel_file = pd.ExcelFile(rs)

print(f"Número total de hojas: {len(excel_file.sheet_names)}\n")

for hoja in excel_file.sheet_names:
    df = pd.read_excel(rs, sheet_name=hoja)
    filas, columnas = df.shape
    print(f"\n Hoja: {hoja} -> Filas: {filas}, Columnas: {columnas}")
    
    # Porcentaje de NA
    porcentaje_na = df.isnull().mean() * 100
    
    # Porcentaje de vacíos (strings vacíos o solo espacios) usando map en lugar de applymap
    porcentaje_vacios = (
        df.apply(lambda col: col.map(lambda x: isinstance(x, str) and x.strip() == ""))
    ).mean() * 100
    
    # Unir ambos resultados en un DataFrame
    resumen = pd.DataFrame({
        '% NA': porcentaje_na,
        '% Vacíos': porcentaje_vacios
    })
    
    # Mostrar todo sin truncar
    print(resumen.to_string())
```

#### Sumatorio de nulos y vacíos para cada hoja

```python
import pandas as pd

# Ruta del archivo Excel
rs = 'DataSET_SF.xlsx'
excel_file = pd.ExcelFile(rs)

print(f"Número total de hojas: {len(excel_file.sheet_names)}\n")

for hoja in excel_file.sheet_names:
    df = pd.read_excel(rs, sheet_name=hoja)
    filas, columnas = df.shape
    
    # Porcentaje de NA por columna
    porcentaje_na = df.isnull().mean() * 100
    
    # Porcentaje de vacíos por columna (usando map para evitar warnings)
    porcentaje_vacios = (
        df.apply(lambda col: col.map(lambda x: isinstance(x, str) and x.strip() == ""))
    ).mean() * 100
    
    # Calcular medias para la hoja
    media_na = porcentaje_na.mean()
    media_vacios = porcentaje_vacios.mean()
    media_total_faltantes = (porcentaje_na + porcentaje_vacios).mean()
    
    # Mostrar resumen compacto
    print(f"Hoja: {hoja} -> Filas: {filas}, Columnas: {columnas}")
    print(f"Media % NA: {media_na:.2f}%")
    print(f"Media % Vacíos: {media_vacios:.2f}%")
    print(f"Media % Datos faltantes (NA + Vacíos): {media_total_faltantes:.2f}%\n")
```

#### Columnas que superan un cierto umbral de datos faltantes

```python
import pandas as pd

# Ruta del archivo Excel
rs = 'DataSET_SF.xlsx'
excel_file = pd.ExcelFile(rs)

print(f"Número total de hojas: {len(excel_file.sheet_names)}\n")

for hoja in excel_file.sheet_names:
    df = pd.read_excel(rs, sheet_name=hoja)
    filas, columnas = df.shape
    
    # Porcentaje de NA por columna
    porcentaje_na = df.isnull().mean() * 100
    
    # Porcentaje de vacíos por columna (usando map para evitar warnings)
    porcentaje_vacios = (
        df.apply(lambda col: col.map(lambda x: isinstance(x, str) and x.strip() == ""))
    ).mean() * 100
    
    # Porcentaje total de datos faltantes (NA + Vacíos)
    porcentaje_total_faltantes = porcentaje_na + porcentaje_vacios
    
    # Medias para la hoja
    media_na = porcentaje_na.mean()
    media_vacios = porcentaje_vacios.mean()
    media_total_faltantes = porcentaje_total_faltantes.mean()
    
    # Columnas con más del 75% de datos faltantes
    cols_mas_75 = (porcentaje_total_faltantes > 75).sum()
    
    # Columnas con el 100% de datos faltantes
    cols_100 = (porcentaje_total_faltantes == 100).sum()
    
    # Mostrar resumen
    print(f"Hoja: {hoja} -> Filas: {filas}, Columnas: {columnas}")
    print(f"Media % NA: {media_na:.2f}%")
    print(f"Media % Vacíos: {media_vacios:.2f}%")
    print(f"Media % Datos faltantes (NA + Vacíos): {media_total_faltantes:.2f}%")
    print(f"Columnas con >75% datos faltantes: {cols_mas_75}")
    print(f"Columnas con 100% datos faltantes: {cols_100}\n")
```

## Ilustración de los datos

#### Media de NA por hoja

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ruta del archivo Excel
rs = 'DataSET_SF.xlsx'  # Cambia por el nombre real de tu archivo
excel_file = pd.ExcelFile(rs)

# Lista para guardar resultados
resultados = []

for hoja in excel_file.sheet_names:
    df = pd.read_excel(rs, sheet_name=hoja)
    
    # Porcentaje de NA por columna
    porcentaje_na = df.isnull().mean() * 100
    
    # Guardar media % NA para la hoja
    resultados.append({
        'Hoja': hoja,
        'Media % NA': porcentaje_na.mean()
    })

# Convertir resultados a DataFrame
df_resultados = pd.DataFrame(resultados)

# Gráfico de barras
plt.figure(figsize=(10,6))
plt.bar(df_resultados['Hoja'], df_resultados['Media % NA'], color='green', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Porcentaje (%)')
plt.ylim(0, 80)  # 🔹 Fijar eje Y de 0 a 100
plt.yticks(range(0, 101, 10))  # 🔹 Marcas de 10 en 10
plt.title('Media de % NA por hoja')
plt.tight_layout()
plt.show()
```

#### Columnas con >75% NA y columnas con 100% NA por hoja

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ruta del archivo Excel
rs = 'DataSet_SF.xlsx'  # Cambia por el nombre real de tu archivo
excel_file = pd.ExcelFile(rs)

# Lista para guardar resultados
resultados = []

for hoja in excel_file.sheet_names:
    df = pd.read_excel(rs, sheet_name=hoja)
    
    # Porcentaje de NA por columna
    porcentaje_na = df.isnull().mean() * 100
    
    # Columnas con más del 75% de datos faltantes
    cols_mas_75 = (porcentaje_na > 75).sum()
    
    # Columnas con el 100% de datos faltantes
    cols_100 = (porcentaje_na == 100).sum()
    
    resultados.append({
        'Hoja': hoja,
        'Cols >75% NA': cols_mas_75,
        'Cols 100% NA': cols_100
    })

# Convertir resultados a DataFrame
df_resultados = pd.DataFrame(resultados)

# Gráfico de barras agrupadas
plt.figure(figsize=(10,6))
bar_width = 0.35
x = range(len(df_resultados))

plt.bar(x, df_resultados['Cols >75% NA'], width=bar_width, label='>75% NA', color='orange', alpha=0.7)
plt.bar([i + bar_width for i in x], df_resultados['Cols 100% NA'], width=bar_width, label='100% NA', color='red', alpha=0.7)

plt.xticks([i + bar_width/2 for i in x], df_resultados['Hoja'], rotation=45, ha='right')
plt.ylabel('Número de columnas')
plt.title('Columnas con >75% NA y 100% NA por hoja')
plt.legend()
plt.tight_layout()
plt.show()
```
